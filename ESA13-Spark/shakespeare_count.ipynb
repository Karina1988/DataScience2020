```python
def cleanInput(lines):
    lines = lines.lower()
    lines = lines.replace("?", "")
    lines = lines.replace("!", "")
    lines = lines.replace(":", "")
    lines = lines.replace(".", "")
    lines = lines.replace(",", "")
    lines = lines.replace("[", "")
    lines = lines.replace("]", "")
    lines = lines.replace(";", "")
    lines = lines.replace("'", "")
    lines = lines.replace('"', "")
    lines = lines.split()
    return lines

lines = sc.textFile("t8.shakespeare.txt")
words = lines.flatMap(cleanInput)

wordCounts = words.countByValue()
sortedWordCounts = sorted(wordCounts.items(), key=lambda x: x[1], reverse=True)

i = 1
for word, count in sortedWordCounts:
    print(f"{i}: {word} - {count} ")
    i+=1
```
