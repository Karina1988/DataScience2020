```python
def cleanInput(lines):
    lines = lines.lower()
    lines = lines.replace("?", "")
    lines = lines.replace("!", "")
    lines = lines.replace(":", "")
    lines = lines.replace(".", "")
    lines = lines.replace(",", "")
    lines = lines.replace("[", "")
    lines = lines.replace("]", "")
    lines = lines.replace(";", "")
    lines = lines.replace("'", "")
    lines = lines.replace('"', "")
    lines = lines.split()
    return lines

lines = sc.textFile("t8.shakespeare.txt")
words = lines.flatMap(cleanInput)

wordCounts = words.countByValue()
sortedWordCounts = sorted(wordCounts.items(), key=lambda x: x[1], reverse=True)

for i in range(100):
    word, count = sortedWordCounts[i]
    print(f"{i+1}: {word} - {count} ")
```
