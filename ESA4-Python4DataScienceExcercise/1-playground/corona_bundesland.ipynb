```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('corona_bundesland.csv', parse_dates=['Tag'])
df= df.sort_values(by=['Tag'], ascending=True)
df
```

```python
ax = plt.gca()
df.plot(kind='line',x='Tag', y='HH', ax=ax)
df.plot(kind='line',x='Tag', y='SH', color='red', ax=ax)
plt.show()
```
