Letâ€™s try to build a classifier to detect the Iris-__Virginica__ type based only on the petal width feature.

```python
%matplotlib inline
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
# http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
iris = datasets.load_iris()
list(iris.keys())
```

```python
is_virginica = (iris["target"] == 2).astype(np.int)  # array of 1 if Iris-Virginica, else 0
```

```python
petal_length = iris["data"][:, (0,)]
fit0 = LogisticRegression()
fit0.fit(petal_length, is_virginica)
X_new = np.linspace(2, 7, 1000).reshape(-1, 1)
y_proba = fit0.predict_proba(X_new)
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")
print(f"probability iris_virginica: {fit0.predict_proba([[4.8]])[:,1]}")
```

```python
petal_width = iris["data"][:, (1,)]
fit1 = LogisticRegression()
fit1.fit(petal_width, is_virginica)
X_new = np.linspace(0, 5, 1000).reshape(-1, 1)
y_proba = fit1.predict_proba(X_new)
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")
print(f"probability iris_virginica: {fit1.predict_proba([[2.5]])[:,1]}")
```

```python
sepal_length = iris["data"][:, (2,)]
fit2 = LogisticRegression()
fit2.fit(sepal_length, is_virginica)
X_new = np.linspace(3.5, 7, 1000).reshape(-1, 1)
y_proba = fit2.predict_proba(X_new)
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")
print(f"probability iris_virginica: {fit2.predict_proba([[5.3]])[:,1]}")
```

```python
sepal_width = iris["data"][:, (3,)]
fit3 = LogisticRegression()
fit3.fit(sepal_width, is_virginica)
X_new = np.linspace(0, 4, 1000).reshape(-1, 1)
y_proba = fit3.predict_proba(X_new)
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")
print(f"probability iris_virginica: {fit3.predict_proba([[2.4]])[:,1]}")
```

```python
all_data = iris["data"]
log_reg = LogisticRegression()
log_reg.fit(all_data, is_virginica)
flower = [[4.8,2.5,5.3,2.4]]
print(f"probability iris_virginica from a combination of all features: {log_reg.predict_proba(flower)[:,1]}")
```
